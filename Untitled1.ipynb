```python
import cv2
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def show(img):
    if (len(img.shape) == 2):
        plt.imshow(sml, cmap='gray')
    else:
        plt.imshow(big)
```

```python
big = mpimg.imread('big.jpeg')
show(big)
```

```python
sml = cv2.imread('sml.png')
sml = cv2.cvtColor(sml, cv2.COLOR_RGB2GRAY)
show(sml)
```

```python
sml_w, sml_h = sml.shape
big_w, big_h,_ = big.shape
```

```python
def img2fv(img):
    if (len(img.shape) == 2):
        return img.reshape(img.shape[0] * img.shape[1])
    else:
        return img.reshape(img.shape[0] * img.shape[1] * img.shape[2])
```

```python
def fv2img(fv,h, w, gray = False):
    if (gray):
        return np.reshape(fv, (w, h))
    else:
        return np.reshape(fv, (w, h, 3))
    
```

```python
#recover test
show(fv2img(img2fv(big), big_h, big_w))
```

```python
#recover test
show(fv2img(img2fv(sml), sml_h, sml_w, True))
```

```python
sml_fv = img2fv(sml)
big_fv = img2fv(big)
result_fv = np.empty(len(big_fv), dtype=int)
```

```python
print((big_w*big_h*3) / (sml_w * sml_h) > 8)
```

```python
def hide_num(tar, offset):
    for i in range(8):
        shifted = tar >> i
        index = offset + i
        if (shifted % 2 == 0):
            if (big_fv[index] % 2 != 0):
                result_fv[index] = big_fv[index] + 1
            else:
                result_fv[index] = big_fv[index]
```

```python
offset = 0
for tar in sml_fv:
    hide_num(tar, offset)
    offset += 8
```

```python
show(fv2img(result_fv, big_h, big_w))
```

```python
recovered_fv = np.empty(len(sml_fv), dtype=int)
```

```python
def recover_num(tar, offset):
    num = 0
    for i in range(8):
        index = offset + i
        if (result_fv[index] % 2 != 0):
            num += pow(2, i)
    return num
```

```python
offset = 0
for i, tar in enumerate(sml_fv):
    recovered_fv[i] = recover_num(tar, offset)
    offset += 8
```

```python
show(fv2img(recovered_fv, sml_h, sml_w, True))
```
